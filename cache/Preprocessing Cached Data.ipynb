```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("pred_const_ip_cache.csv")
df = df[~df["RegionName"].isin(df["RegionName"].unique()[1:])].reset_index(drop=True)
df = df.rename(columns={"PredictedDailyNewCases":"DailyNewCases"})
df["DailyNewCasesMA"] = df["DailyNewCases"].rolling(7).mean()
df["DailyNewCasesMA"].fillna(df["DailyNewCases"], inplace=True)
df["DailyNewCases"] = df["DailyNewCasesMA"]
df = df.sort_values(by="CountryName").reset_index(drop=True)
df = df.drop(columns=["DailyNewCasesMA"])
```
